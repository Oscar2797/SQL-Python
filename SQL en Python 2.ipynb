```python
import sqlite3
conn= sqlite3.connect("Clientes2.db")
```

```python
cur=conn.cursor()
```

```python
# Lectura de archivo CSV

import os
import csv
os.chdir("C:/Users/odela/Downloads")
with open("Cientifico de datos M40 Customers.csv") as f:
    reader=csv.reader(f)
    data=list(reader)
```

```python
data
```

```python
len(data)
```

```python
# Creación de la tabla donde se importará el archivo CSV

cur.execute("""
    CREATE TABLE customers(
     CustomerID REAL PRIMARY KEY,
        CustomerName TEXT,
        ContactName TEXT,
        Address TEXT,
        City TEXT,
        PostalCode VARCHAR TEXT,
        Country TEXT,
        Purchase REAL
        );
""")
```

```python
# Inserción de los datos en la tabla recien creada

for renglon in range(1,len(data)):
    cur.execute("""
        INSERT INTO customers(CustomerID,CustomerName , ContactName,Address,City,PostalCode,Country,Purchase)
        VALUES(?,?,?,?,?,?,?,?)
    """,data[renglon])
```

```python
# Aplicación de los cambios

conn.commit()
```
