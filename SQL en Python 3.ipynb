```python
import os
import csv
os.chdir("C:/Users/odela/Downloads")
with open("Cientifico de datos M40 Customers.csv") as f:
    reader=csv.reader(f)
    data=list(reader)
data
```

```python
cur.execute("""
    CREATE TABLE customers2(
     CustomerID REAL PRIMARY KEY,
        CustomerName TEXT,
        ContactName TEXT,
        Address TEXT,
        City TEXT,
        PostalCode VARCHAR TEXT,
        Country TEXT,
        Purchase REAL
        );
""")
```

```python
for renglon in range(1,len(data)):
    cur.execute("""
        INSERT INTO customers2(CustomerID,CustomerName , ContactName,Address,City,PostalCode,Country,Purchase)
        VALUES(?,?,?,?,?,?,?,?)
    """,data[renglon])
```

```python
# Despliegue ordenado de registros de acuerdo a un criterio 

cur.execute("SELECT CustomerName, City FROM customers2 ORDER BY City;")
result= cur.fetchall()
result
```

```python
# Despliegue ordenado de registros de acuerdo a un criterio explícito en orden ascendente
cur.execute("SELECT ContactName, Country FROM customers2 ORDER BY ContactName ASC;")
result= cur.fetchall()
result
```

```python
# Despliegue ordenado de registros de acuerdo a un criterio explícito en orden descendente
cur.execute("SELECT ContactName, Country FROM customers2 ORDER BY ContactName DESC;")
result= cur.fetchall()
result
```

```python
# Despliegue ordenado de registros de acuerdo a un criterio de orden descendente sin valores nulos
cur.execute("SELECT ContactName, Country FROM customers2 WHERE Country IS NOT NULL ORDER BY ContactName DESC;")
result= cur.fetchall()
result
```

```python
# Agrupación por campo individual
cur.execute("SELECT Country, Count(ContactName) FROM customers2  GROUP BY Country ;")
result= cur.fetchall()
result
```

```python
# Agrupación por campos multiples
cur.execute("SELECT Country,PostalCode, Count(ContactName) FROM customers2  GROUP BY Country, PostalCode ;")
result= cur.fetchall()
result
```

```python
# el primer criterio que toma es Count(ContactName)
# Uso de agrupamientos con ordenamientos     AS Contactos es un alias que se le da y que posteriormente se puede llamar, en este caso es el alias para  Count(ContactName)
cur.execute("""SELECT Country,PostalCode, Count(ContactName) AS Contactos FROM customers2 GROUP BY Country, PostalCode
                ORDER BY Contactos;""")
result= cur.fetchall()
result
```

```python
# Forma alternatva con comodines
cur.execute("""SELECT Country,PostalCode, Count(*) AS Contactos FROM customers2 GROUP BY Country, PostalCode
                ORDER BY Contactos;""")
result= cur.fetchall()
result
```

```python
# Consulta con límites de registro
cur.execute("""SELECT Country,PostalCode, Count(*) AS Contactos FROM customers2 GROUP BY Country, PostalCode
                ORDER BY Contactos LIMIT 5;""")
result= cur.fetchall()
result
```

```python
# filtros para datos agrupados (Count, sum, avg, min, etc)
cur.execute("SELECT Country,Count(ContactName)  FROM customers2 GROUP BY Country HAVING  Count(ContactName)>1 ;")
result= cur.fetchall()
result
```

```python
# Selección de paises específicos 
cur.execute("""SELECT Country,Count(ContactName)  FROM customers2  WHERE Country IN ('Austria', 'Germany','USA')
            GROUP BY Country HAVING Count(ContactName)>3 ;""")
result= cur.fetchall()
result
```
