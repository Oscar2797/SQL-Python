```python
# Lectura de archvo csv

import os
import csv

os.chdir("C:/Users/odela/Downloads")
with open("Customers1.csv") as f:
    reader= csv.reader(f)
    data=list(reader)
data
```

```python
# Revisi√≥n tabla 1

cur.execute("SELECT * FROM Customers1 LIMIT 10;")
result= cur.fetchall()
result
```

```python
# Revisi√≥n tabla 2

cur.execute("SELECT * FROM Customers2 LIMIT 10;")
result= cur.fetchall()
result
```

```python
# Consutla mediante INNER JOIN

cur.execute( """ SELECT p1.CustomerName, p1.City, p2.ContactName, p2.PurchaseUSD 
                 FROM Customers1 AS p1 INNER JOIN Customers2 AS p2
                 ON p1.CustomerName= p2.CustomerName;""")
result= cur.fetchall()
result
```

```python
# OTRA ALTERNATIVA (NO REQUIERE DEFINIR TABLA 2)

cur.execute( """ SELECT p1.CustomerName, p1.City, ContactName, PurchaseUSD 
                 FROM Customers1 AS p1 INNER JOIN Customers2 AS p2
                 ON p1.CustomerName= p2.CustomerName;""")
result= cur.fetchall()
result
```

```python
# OTRA ALTERNATIVA MEDIANTE COMANDO USING (S√ìLO APLICA CUANDO AMBAS TABLAS TIENEN EL MISMO NOMBRE EN EL CAMPO EN COM√öN, en este caso CustomerName)

cur.execute("""SELECT p1.CustomerName, p1.City, CustomerName, PurchaseUSD
               FROM Customers1 AS p1 INNER JOIN Customers2 AS p2
               USING (CustomerName);""")
result= cur.fetchall()
result
```

```python
#  CONSULTA MEDIANTE LEFT JOIN

cur.execute( """ SELECT p1.CustomerName, p1.City, p2.ContactName, p2.PurchaseUSD 
                 FROM Customers1 AS p1 LEFT JOIN Customers2 AS p2
                 ON p1.CustomerName= p2.CustomerName;""")
result= cur.fetchall()
result
```

```python
#  USO DE CASE WHEN para separar por clases los grupos de compra (END AS define el nombre de la columna que se crea)

cur.execute("""SELECT CustomerName, ContactName,
               CASE WHEN PurchaseUSD< 10000 THEN "Venta baja"
               WHEN PurchaseUSD<= 15000 THEN "Venta media"
               ELSE "Venta alta" END AS Grupo_Venta
               FROM Customers2
               ORDER BY Grupo_Venta;""")
result= cur.fetchall()
result
```

```python
# Uso de operai√≥n UNION (No repite registros)

cur.execute(""" SELECT CustomerName, ContactName, PostalCode, PurchaseUSD
                FROM Customers2
                UNION 
                SELECT CustomerName, ContactName, PostalCode, PurchaseUSD
                FROM Customers3
                ORDER BY CustomerName;
            """)
result= cur.fetchall()
result
```

```python
# Uso de operai√≥n UNION ALL (S√≠ repite registros)

cur.execute(""" SELECT CustomerName, ContactName, PostalCode, PurchaseUSD
                FROM Customers2
                UNION ALL
                SELECT CustomerName, ContactName, PostalCode, PurchaseUSD
                FROM Customers3
                ORDER BY CustomerName;
            """)
result= cur.fetchall()
result
```

```python
# uso de la operaci√≥n INTERSECT (S√≥lo muestra registros comunes)

cur.execute(""" SELECT CustomerName, ContactName, PostalCode, PurchaseUSD
                FROM Customers2
                INTERSECT
                SELECT CustomerName, ContactName, PostalCode, PurchaseUSD
                FROM Customers3
                ORDER BY CustomerName;
            """)
result= cur.fetchall()
result
```

```python
# uso de la operaci√≥n EXCEPT (Muestra los registros de la primer tabla y que no est√°n en la segunda)

cur.execute(""" SELECT CustomerName, ContactName, PostalCode, PurchaseUSD
                FROM Customers2
                EXCEPT
                SELECT CustomerName, ContactName, PostalCode, PurchaseUSD
                FROM Customers3
                ORDER BY CustomerName;
            """)
result= cur.fetchall()
result
```

```python
# SEMI JOIN Ejemplo ( datos que est√°n dentro de condici√≥n de un asegunda tabla)

cur.execute(""" SELECT CustomerName, Country
                FROM Customers1
                WHERE CustomerName IN
                    (SELECT CustomerName
                     FROM Customers2
                     WHERE PurchaseUSD<10000);
            """)
result= cur.fetchall()
result
```

```python
# ANTI JOIN Ejemplo (datos que no est√°n en condici√≥n de una segunda tabla)

cur.execute(""" SELECT CustomerName, Country
                FROM Customers1
                WHERE CustomerName NOT IN
                    (SELECT CustomerName
                     FROM Customers2
                     WHERE PurchaseUSD<10000);
            """)
result= cur.fetchall()
result
```

```python
# Uso de subquery para seleccionar clientes cuyo monto de compra sea menor al promedio general

cur.execute(""" SELECT CustomerName, PurchaseUSD
                FROM Customers2
                WHERE PurchaseUSD <
                    (SELECT AVG(PurchaseUSD) FROM Customers2);
            """)
result= cur.fetchall()
result
```
